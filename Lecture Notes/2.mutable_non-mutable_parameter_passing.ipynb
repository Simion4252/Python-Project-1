## Passing by Refrence -- Mutable Object
- List
- Dictionary
- Set

```python
def edit_mutable(obj, kind): # we have two formal parameters obj,kind
    
    print(f"Memory Adress in  the func: {id(obj)}")
    
    if kind == "list":    
        obj.append("d")
    elif kind == "dict":
        obj[4] = "d"
    elif kind == "set":
        obj.update("d")
    else:
        print("Value of 'kind' is Unknow!")
    print(f"Object in  the func: {obj}")
```

```python
liste = ["a", "b", "c"]
print(f"Object out the func: {liste}")
print(f"Memory Adress out the func: {id(liste)}")
edit_mutable(liste, kind="list")
print(f"Object out the func: {liste}")
```

```python
dic = {1:"a", 2:"b", 3:"c"}
print(f"Object out the func: {dic}")
print(f"Memory Adress out the func: {id(dic)}")
edit_mutable(dic, kind="dict")
print(f"Object out the func: {dic}")
```

```python
ens = {"a", "b", "c"}
print(f"Object out the func: {ens}")
print(f"Memory Adress out func: {id(ens)}")
edit_mutable(ens, kind="set")
print(f"Object out the func: {ens}")
```

## Passing br value - Immutable

```python
def edit_unmutable(obj, kind):
    
    print(f"Memory Adress in  the func: {id(obj)}")
    
    if kind == "int":    
        obj = 0
    elif kind == "float":    
        obj = 0.0
    elif kind == "bool":    
        obj = True
    elif kind == "string":
        obj = "dev inst"
    elif kind == "tuple":
        obj = ("a", "b", "c", "d")
    else:
        print("Value of 'kind' is Unknow!")
    print(f"Object in  the func: {obj}")
```

```python
x = 2024
print(f"Object out the func: {x}")
print(f"Memory Adress out the func: {id(x)}")
edit_unmutable(x, kind="int")
print(f"Object out the func: {x}")
```

```python
y = 20.24
print(f"Object out the func: {y}")
print(f"Memory Adress out the func: {id(y)}")
edit_unmutable(y, kind="float")
print(f"Object out the func: {y}")
```

```python
bol = False
print(f"Object out the func: {bol}")
print(f"Memory Adress out the func: {id(bol)}")
edit_unmutable(bol, kind="bool")
print(f"Object out the func: {bol}")
```

```python
ch = "developpers intitute"
print(f"Object out the func: {ch}")
print(f"Memory Adress out the func: {id(ch)}")
edit_unmutable(ch, kind="string")
print(f"Object out the func: {ch}")
```

```python
tup = ("a", "b", "c")
print(f"Object out the func: {tup}")
print(f"Memory Adress out the func: {id(tup)}")
edit_unmutable(tup, kind="tuple")
print(f"Object out the func: {tup}")
```
